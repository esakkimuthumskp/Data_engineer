## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
df1 = spark.read.option("multiline","true").format("json").load("dbfs:/FileStore/shared_uploads/Ceena.Christudhas@agilisium.com/insta_datas2___1_.json")
df1.display()
```

```
temp_table_name = "insta_datas2_json"
df1.createOrReplaceTempView(temp_table_name)
```

```
%sql
select category,product_type,count(product_type),round(avg(like_count)) as like_count from insta_datas2_json where category = "Gopinath" group by category,product_type
```

```
%sql
select category,round(avg(like_count)) as likes, round(avg(view_count)) as views, round(avg(comment_count)) as comment_count from insta_datas2_json group by category
```

```
df1=spark.sql("select product_type,round(avg(like_count)) as likes_for_dur_lessthan_30 from insta_datas2_json where video_duration < 30 and category=\"Gopinath\" group by product_type")
df1.display()
```

```
df2=spark.sql("select product_type,round(avg(like_count)) as likes_for_dur_btw_30to60 from insta_datas2_json where video_duration between 30 and 60 and category=\"Gopinath\" group by product_type")
df2.display()
```

```
df3=spark.sql("select product_type,round(avg(like_count)) as likes_for_dur_abv_60 from insta_datas2_json where video_duration>60 and category=\"Gopinath\" group by product_type")
df3.display()
```

```
j1 = df1.join(df2,on=['product_type'],how='left')
j1.display()
```

```
j2 = j1.join(df3,on=['product_type'],how='left')
j2.display()
```

```
%sql
select category,product_type,count(product_type) products,round(avg(like_count)) as like_count from insta_datas2_json where category = "Gopinath" group by category,product_type
```

```
# File location and type
file_location = "/FileStore/tables/insta_datas_3.json"
file_type = "json"

# CSV options
infer_schema = "true"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option('multiline','true') \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create a view or table

temp_table_name = "insta_datas_3_json"

df.createOrReplaceTempView(temp_table_name)
```

```
# With this registered as a temp view, it will only be available to this particular notebook. If you'd like other users to be able to query this table, you can also create a table from the DataFrame.
# Once saved, this table will persist across cluster restarts as well as allow various users across different notebooks to query this data.
# To do so, choose your table name and uncomment the bottom line.

permanent_table_name = "insta_datas_3_json"

# df.write.format("parquet").saveAsTable(permanent_table_name)
```
