```python
def Load_Excel_extract_json(file_name,name=''):
    Link_list = pd.read_excel(file_name)
    Json_list =list()
    for x in Link_list.columns:
        Column_data = Link_list[x]     
        for y in Column_data:
            if type(y)==type(''):
                Json=extract_json(y)
                try:
                    Slice_json = dict(Json['items'][0])
                    Slice_json['category']=x
                    Slice_json['name'] = name
                    Json_list.append(Slice_json)
                except:
                    Json_list = 'Now not available API'
                    break
                    
                
    return Json_list
```

```python
def get_Specific_data(Json_list):
    Specific_data=list()
    for n in Json_list:
        Temp_Dict = {}
        Temp_Dict['category'] = n['category']
        Temp_Dict['name'] = n['name']
        Temp_Dict['video_id'] = n['id']
        Temp_Dict['video_link'] = n['code']
        Temp_Dict['caption'] = n['caption']['text']
        Temp_Dict['media_type'] = n['media_type']
        Temp_Dict['product_type'] =n['product_type']
        Temp_Dict['user_id'] = n['caption']['user_id']
        Temp_Dict['user_name'] = n['user']['username']
        try:
            Temp_Dict['title'] = n['title']
        except:
            pass
        try:
            Temp_Dict['video_duration'] = n['video_duration']
        except:
            Temp_Dict['video_duration'] = 0
        try:
            Temp_Dict['view_count'] = n['view_count']
        except:
            Temp_Dict['view_count'] = 0
        try:
            Temp_Dict['play_count'] = n['play_count']
        except:
            Temp_Dict['play_count'] = 0
        Temp_Dict['like_count'] = n['like_count']
        Temp_Dict['comment_count'] = n['comment_count']
        Specific_data.append(Temp_Dict)
        del Temp_Dict
    return Specific_data
```
